# Calculos nominais
## Circuito 1

```python
E = 5.0 # [V]
R_1 = 1000. # [Ohms]
R_2 = 2000. # [Ohms]

R_eq = R_1+R_2

i_1 = E/R_eq

V_2 = R_1*i_1

V_1 = R_2*i_1

P_1 = R_1*(i_1**2)
P_2 = R_2*(i_1**2)

print(f"R_eq = {R_eq}")
print(f"I_1 = {i_1}")
print(f"V_1 = {V_1} V")
print(f"V_2 = {V_2} V")
print(f"P_1 = {P_1} W")
print(f"P_2 = {P_2} W")
```
